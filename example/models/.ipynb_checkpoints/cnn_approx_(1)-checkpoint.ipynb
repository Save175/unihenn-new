```python
import os
import random
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import shutil

# Function to split dataset into train, validation, and test sets
def split_data(source_folder, dest_folder, ratio=(0.8, 0.1, 0.1)):
    for category in os.listdir(source_folder):
        category_path = os.path.join(source_folder, category)
        if not os.path.isdir(category_path):
            continue
        images = os.listdir(category_path)
        random.shuffle(images)
        total = len(images)
        train_count = int(total * ratio[0])
        val_count = int(total * ratio[1])

        os.makedirs(os.path.join(dest_folder, 'train', category), exist_ok=True)
        os.makedirs(os.path.join(dest_folder, 'val', category), exist_ok=True)
        os.makedirs(os.path.join(dest_folder, 'test', category), exist_ok=True)

        for i, img in enumerate(images):
            img_path = os.path.join(category_path, img)
            if i < train_count:
                shutil.copy(img_path, os.path.join(dest_folder, 'train', category))
            elif i < train_count + val_count:
                shutil.copy(img_path, os.path.join(dest_folder, 'val', category))
            else:
                shutil.copy(img_path, os.path.join(dest_folder, 'test', category))

# Custom dataset class
class CustomDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.images = []
        self.labels = []
        
        # Create a mapping from class names to binary labels
        self.label_map = {label: idx for idx, label in enumerate(sorted(os.listdir(root_dir)))}

        for label in os.listdir(root_dir):
            label_dir = os.path.join(root_dir, label)
            if not os.path.isdir(label_dir):
                continue
            for img in os.listdir(label_dir):
                self.images.append(os.path.join(label_dir, img))
                self.labels.append(self.label_map[label])  # Use mapped binary label

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

# Transform function including normalization
def transform_image(image):
    # Resize to 100x100
    image = image.resize((100, 100))
    image = np.array(image)  # Convert to numpy array
    image = image / 255.0  # Normalize to [0, 1]

    # Apply mean and std normalization
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image = (image - mean) / std  # Normalize to mean 0 and std 1
    
    image = torch.tensor(image, dtype=torch.float32)  # Convert to tensor
    image = image.permute(2, 0, 1)  # Change to CxHxW format
    return image

# Split the dataset
file_path = './concrete_crack_dataset'
splitted_folder = 'data'
split_data(file_path, splitted_folder)

# Create datasets
train_dataset = CustomDataset(os.path.join(splitted_folder, 'train'), transform=transform_image)
val_dataset = CustomDataset(os.path.join(splitted_folder, 'val'), transform=transform_image)
test_dataset = CustomDataset(os.path.join(splitted_folder, 'test'), transform=transform_image)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Example usage: Print shapes of the first batch
for images, labels in train_loader:
    print(f"Image batch shape: {images.shape}")
    print(f"Labels: {labels}")
    break
```

```python
def _approximated_ReLU(x):
    return 0.117071 * x**2 + 0.5 * x + 0.375373

class Square(torch.nn.Module):
    def forward(self, x):
        return x**2

class ApproxReLU(torch.nn.Module):
    def forward(self, x):
        return _approximated_ReLU(x)

class Flatten(torch.nn.Module):
    def forward(self, x):
        return torch.flatten(x, 1)
```

```python
class ElementWiseMultiply(torch.nn.Module):
    def forward(self, x1, x2):
        return torch.mul(x1,x2)
```

```python
import torch.nn as nn
class M15(nn.Module):
    def __init__(self):
        super(M15, self).__init__()

        # Convolutional Layers
        self.conv1 = nn.Conv2d(3, 8, kernel_size=3, padding=1)
        self.approx_relu1 = ApproxReLU()
        self.avgpool1 = nn.AvgPool2d(2, 2)

        self.conv2 = nn.Conv2d(8, 16, kernel_size=3, padding=1)
        self.approx_relu2 = ApproxReLU()
        self.avgpool2 = nn.AvgPool2d(2, 2)

        # Spatial attention convolution
        self.spatial_conv = nn.Conv2d(16, 1, kernel_size=3, padding=1, bias=False)

        # Element-wise multiplication layer
        self.element_wise_mul = ElementWiseMultiply()

        # Fully connected layers
        self.flatten = Flatten()
        self.fc1 = nn.Linear(16 * 25 * 25, 128)  # Updated based on calculated flattened output
        self.approx_relu3 = ApproxReLU()
        self.fc2 = nn.Linear(128, 2)

        # Activation for spatial attention
        self.square_spatial = Square()

    def forward(self, x):
        # Forward through conv1
        x = self.conv1(x)
        x = self.approx_relu1(x)
        x = self.avgpool1(x)

        # Forward through conv2
        x = self.conv2(x)
        x = self.approx_relu2(x)
        x = self.avgpool2(x)

        # Spatial Attention Mechanism
        spatial_attention = self.spatial_conv(x)  # Convolution for spatial attention
        spatial_attention = self.square_spatial(spatial_attention)  # Square activation

        # Element-wise multiplication of the input with the spatial attention
        x = self.element_wise_mul(x, spatial_attention)

        # Flatten and fully connected layers
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.approx_relu3(x)
        x = self.fc2(x)

        return x
```

```python
class M15(nn.Module):
    def __init__(self):
        super(M15, self).__init__()

        # Convolutional Layers
        self.conv1 = nn.Conv2d(3, 8, kernel_size=3, padding=1)
        self.approx_relu1 = ApproxReLU()
        self.avgpool1 = nn.AvgPool2d(2, 2)

        self.conv2 = nn.Conv2d(8, 16, kernel_size=3, padding=1)
        self.approx_relu2 = ApproxReLU()
        self.avgpool2 = nn.AvgPool2d(2, 2)

        # Spatial attention convolution
        self.spatial_conv = nn.Conv2d(16, 1, kernel_size=3, padding=1, bias=False)

        # Element-wise multiplication layer
        self.element_wise_mul = ElementWiseMultiply()

        # Fully connected layers
        self.flatten = Flatten()
        self.fc1 = nn.Linear(16 * 25 * 25, 128)  # Assuming input image size is 100x100
        self.approx_relu3 = ApproxReLU()
        self.fc2 = nn.Linear(128, 2)

        # Activation for spatial attention
        self.square_spatial = Square()

    def forward(self, x):
        # Forward through conv1
        x = self.conv1(x)
        x = self.approx_relu1(x)
        x = self.avgpool1(x)

        # Forward through conv2
        x = self.conv2(x)
        x = self.approx_relu2(x)
        x = self.avgpool2(x)
        print(x.shape)

        # Spatial Attention Mechanism
        spatial_attention = self.spatial_conv(x)  # Convolution for spatial attention
        spatial_attention = self.square_spatial(spatial_attention)  # Apply the square activation

        # Check if spatial_attention has the expected shape
        print(f"spatial_attention shape: {spatial_attention.shape}")

        # Expand spatial_attention to match the number of channels in x
        # Here we ensure the spatial dimensions (25x25) are retained and only channels are expanded
        spatial_attention = spatial_attention.expand(-1, x.size(1), -1, -1)  # Expand to match x's channel dimension

        # Check shape after expand
        print(f"spatial_attention after expand shape: {spatial_attention.shape}")

        # Element-wise multiplication
        x = self.element_wise_mul(x, spatial_attention)

        # Flatten and fully connected layers
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.approx_relu3(x)
        x = self.fc2(x)

        return x
```
